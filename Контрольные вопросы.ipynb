```python
import numpy
import pandas
import matplotlib.pyplot
cube=[6, 3, 3]
rubick = numpy.ones(cube)
for k in range(1,6):
    for i in range(0,3):
        for j in range(0,3):
            rubick[(k,i,j)]=k+1
rubick
```
